````python
import csv
import pprint
import math

file_path = 'data.csv'
````

````python
with open(file_path) as f:
    print(f.read())
````

````python
fields = ["name", "style", "hp", "attack", "defense", "sp_atk", "sp_def", "speed", "move1", "move2", "move3", "move4", "held_item"]
types = ["string", "integerl", "integer", "integer", "integer", "integer", "integer", "integer", "string", "string", "string", "string", "string"]
texts = []

def append_debug(array, value):
    print(value)
    array.append(value)
    
def isfloat(string):
    try:
        float(string)  # 文字列をfloatにキャスト
        return True
    except ValueError:
        return False

with open(file_path) as f:
    reader = csv.reader(f)
    pokemon_data = list(reader)[1:]
    for row in pokemon_data:
#         print("Pokemon.create!(")
#         texts.append("Pokemon.create!(")
        append_debug(texts, "Pokemon.create!(")
        for i in range(len(fields)):
            column_text = "  " + fields[i] + ": "
            if i == len(fields) - 1:
                value_text =  '"' + row[i] + '"'
            elif types[i] == "string":
                value_text =  '"' + row[i] + '",'
            elif types[i] == "bool":
                value_text = "true," if (row[i] == "o") else "false,"
            else:
                value_text = str(math.floor(float(row[i]))) + ','
            text = column_text + value_text
#             print(text)
#             texts.append(text)
            append_debug(texts, text)
#         print(")")
#         texts.append(")")
        append_debug(texts, ")")
#         print("")
#         texts.append("")
        append_debug(texts, "")
````
